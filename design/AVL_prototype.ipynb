```python
import math

x = np.linspace(0, 1000, 1000)
lux = 22
internal_adj = 1000 # [-1 1]
_scale = 0.316 # [-1 .316 1]
offset = 0    # [-1 0 1]

def step0_10(lux):
    out = lux/internal_adj
    out = pow(out, 1./4.)
    scale = _scale
    if scale < 0:
        scale = -pow(scale, 2)
    else:
        scale = pow(scale, 2)
    scale = scale * 10
        
    out = out + offset * 10
    out = out * scale
        
#   retur out
    return out

vstep = np.vectorize(step0_10)


plt.plot(x, vstep(x))
# np.sin()
```
